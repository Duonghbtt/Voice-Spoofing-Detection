```python
import os
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import pandas as pd

base_npy_2019 = r'D:\DL\output_npy_2019'
base_npy_2021 = r'D:\DL\output_npy_2021'
```

```python
label_19 = pd.read_csv(os.path.join(base_npy_2019, 'output_mfcc', 'labels_train.csv'))
label_21 = pd.read_csv(os.path.join(base_npy_2021, 'output_mfcc', 'labels_eval_2021.csv'))

bonafide_19 = label_19[label_19['label'] == 1].iloc[0]['filename']
bonafide_21 = label_21[label_21['label'] == 1].iloc[0]['filename']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

lfcc_19 = np.load(os.path.join(base_npy_2019, 'output_lfcc', 'train', bonafide_19 + '.npy'))
img_19 = librosa.display.specshow(lfcc_19, x_axis='time', ax=axes[0], hop_length=160)
axes[0].set_title('2019 BONAFIDE (Moi truong sach) - LFCC')
fig.colorbar(img_19, ax=axes[0])

lfcc_21 = np.load(os.path.join(base_npy_2021, 'output_lfcc', 'eval_2021', bonafide_21 + '.npy'))
img_21 = librosa.display.specshow(lfcc_21, x_axis='time', ax=axes[1], hop_length=160)
axes[1].set_title('2021 BONAFIDE (Nhieu vien thong) - LFCC')
fig.colorbar(img_21, ax=axes[1])

plt.tight_layout()
plt.show()
```
